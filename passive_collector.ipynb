```python
# We have a dictionary of {ipA, ipB} = ['udp', ]

names = ['src', '<->', 'dest', 'in_frames', 'in_bytes', 'out_frames', 'out_bytes', 'total_frames', 'total_bytes', 'rel_start', 'duration']
protocol_names = ['total' + "_" + name for name in names]

passive_files = glob.glob('resources/smallpassive*.csv')
d = collections.defaultdict(list)
for i in range(len(passive_files)):
    result = re.search('_(.*).tshark', passive_files[i])
    protocol = result.group(1)
    if "PORTS" in protocol:
        continue
    passive_df = pd.read_csv(passive_files[i], header=None, skiprows=1, names=protocol_names)
    for i, row in passive_df.iterrows():
        if row.out_frames > 0:
            if row.duration != 0:
                d[row.src, row.d][protocol] = row.out_bytes/row.duration, row.out_frames/row.duration
#                 d[row.src, row.dest].append([protocol, row.out_bytes/row.duration, row.out_frames/row.duration])
        if row.in_frames > 0:
            if row.duration != 0:
                d[row.dest, row.src][protocol] = row.in_bytes/row.duration, row.in_frames/row.duration
#                 d[row.dest, row.src].append([protocol, row.in_bytes/row.duration, row.in_frames/row.duration])


print(d)
```
